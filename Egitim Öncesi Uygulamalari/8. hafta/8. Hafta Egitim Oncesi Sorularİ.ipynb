# Problem 3

```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
# 3.Problemin Çözümünü Buraya Yazınız
class Vehicle():
    def __init__(self,name,speed,mileage):
        self.name = name
        self.speed = speed
        self.mileage = mileage
class Bus(Vehicle):
    def __init__(self,color):
        self.color = color
car2 = Bus("blue")

car1 = Bus("Black")
car1.name = "Volvo"
car1.speed = 120
car1.mileage = 12


print(f"Name: {car1.name} Speed: {car1.speed} Mileage: {car1.mileage} Color: {car1.color} ")

````
